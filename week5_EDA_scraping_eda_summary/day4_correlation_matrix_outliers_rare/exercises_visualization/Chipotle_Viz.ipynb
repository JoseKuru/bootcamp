### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# remember to %matplotlib inline
%matplotlib inline
```

### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

### Step 3. See the first 10 entries

```python
chipo.head(10)
```

### Step 4. Create a histogram of the top 5 items bought

```python
# Antes de crear el diagrama de barras creamos un dataframe 'barplot' con los valores ya preparados.
barplot = chipo.groupby('item_name').sum().sort_values('quantity', ascending=False).loc[:, ['quantity']].iloc[:5]
barplot
```

```python
plt.bar(x=barplot.index, height=barplot.quantity,)
plt.title("Most ordererd Chipotle's Items")
plt.xticks(rotation='vertical')
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
```

### Step 5. Create a scatterplot with the number of items orderered per order price

Make sure you get the same labels and title
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
# Cambiamos el tipo de dato de 'item_price' a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
plt.scatter(x=chipo.groupby('order_id').sum()['item_price'], y=chipo.groupby('order_id').sum()['quantity'], c='green')
plt.title('Number of items ordered per order price')
plt.xlabel('order price')
plt.ylabel('items orderes')
```
