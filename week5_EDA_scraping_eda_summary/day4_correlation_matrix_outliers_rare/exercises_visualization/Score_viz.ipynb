### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# remember to %matplotlib inline
%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
data = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 'age': [42, 52, 36, 24, 73], 'female': [0, 1, 1, 0, 1], 'preTestScore': [4, 24, 31, 2, 3], 'postTestScore': [25, 94, 57, 62, 70]})
data
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(x=data.preTestScore, y=data.postTestScore, s=data.age, c='blue')
plt.title('preTestScore x postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
colour = {0: 'blue', 1:  'red'}
plt.scatter(x=data.preTestScore, y=data.postTestScore, s=data.postTestScore*(4.5), c=data.female, cmap='RdBu')
plt.title('preTestScore x postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
```
